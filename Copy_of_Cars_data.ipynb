```python
import matplotlib.pyplot as plt


# Identify categorical columns
categorical_columns = ['make', 'model', 'body_type', 'fuel_system', 'fuel_type',
                       'basic_warranty', 'child_safety_locks', 'parking_assistance',
                       'navigation_system']

# Plot bar charts for each categorical variable
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    new_df[column].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Count of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
